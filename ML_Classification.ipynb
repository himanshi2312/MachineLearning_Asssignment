```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for i in bank_data.select_dtypes('object').columns:
    bank_data[i] = le.fit_transform(bank_data[i])
```

```python
bank_data
```

```python
bank_data.columns
```

```python
X = bank_data.drop([ 'contact', 'day', 'month','pdays','previous','deposit'],axis =1)
```

```python
y = bank_data["deposit"]
```

```python
from sklearn.preprocessing import StandardScaler

st = StandardScaler()
```

```python
X["balance"] = st.fit_transform(X[["balance"]])
X["duration"] = st.fit_transform(X[["duration"]])
```

```python
X.head()
```

## Split Training and Testing bank_data

```python
from sklearn.model_selection import train_test_split

x_train , x_test ,y_train ,y_test =train_test_split(X,y ,test_size=0.3, random_state= 42)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## Model 1 using Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression(max_iter=200)
lg.fit(x_train, y_train)
```

```python
y_pred_lg = lg.predict(x_test)
print('Accuracy of logistic regression classifier on test set: {:.3f}'.format(lg.score(x_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred_lg)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_lg))
```

## Model 2 using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dct = DecisionTreeClassifier()
dct.fit(x_train, y_train)
```

```python
y_pred_dct = dct.predict(x_test)
print('Accuracy of Decision tree classifier on test set: {:.3f}'.format(dct.score(x_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred_dct)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_dct))
```

## Model 3 using KNN 

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
y_pred_knn = knn.predict(x_test)
print('Accuracy of KNN classifier on test set: {:.3f}'.format(knn.score(x_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred_knn)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_knn))
```

## Model 4 using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
```

```python
y_pred_rfc = knn.predict(x_test)
print('Accuracy of Random Forest classifier on test set: {:.3f}'.format(rfc.score(x_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred_rfc)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_rfc))
```

## Model Comparison

```python
print("Logistic Regression model score:",lg.score(x_test, y_test))
print("Decision Tree classifier model score:",dct.score(x_test, y_test))
print("KNN classifier model score:",knn.score(x_test, y_test))
print("Random Forest Classifier model score:",rfc.score(x_test, y_test))
```
